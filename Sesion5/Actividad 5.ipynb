## Limpia los datos y actualiza el archivo CSV.

Vemos los valores perdidos del dataframe

```python
def ver_null(df):
    total = df.isnull().sum().sort_values(ascending = False)
    percent = (df.isnull().sum() / df.isnull().count()).sort_values(ascending = False)
    missing_data = pd.concat([total, percent], axis = 1, keys = ['Total', 'Porcentaje'])
    return missing_data
ver_null(df).head(10)
```

Observamos que hay 3 valores en la columna precio (price) que éstan vacíos. Vamos a rellenarlos con la media de los valores de dicha columna

```python
df_limpio = df.fillna(df.mean())
ver_null(df_limpio)
```

Hemos comprobado que no tiene valores vacíos. Vamos a actualizar el archivo CSV.

```python
df_limpio.to_csv('./Automobile_data_limpio.csv')
```

## Encuentra el nombre de la compañía de automóviles más cara.

```python
company = df_limpio['company'][df_limpio['price']==df_limpio['price'].max()].values[0]
print("La compañía más cara es: {}".format(company))
```
