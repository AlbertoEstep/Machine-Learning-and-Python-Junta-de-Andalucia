```python
import pandas as pd
import numpy as np
import math

r_cols = ['user_id', 'movie_id', 'rating']  
ratings = pd.read_csv("dataset_unido/u.data", sep='\t', names=r_cols, usecols=range(3), encoding="ISO-8859-1")

m_cols = ['movie_id', 'title']  
movies = pd.read_csv('dataset_unido/u.item', sep='|', names=m_cols, usecols=range(2), encoding="ISO-8859-1")

# combinamos ambos datasets para tener el  
ratings = pd.merge(movies, ratings)

# Pivotamos la tabla para que la matriz tenga : fila por usuario y columna por pelicula
movieRatings = ratings.pivot_table(index=['user_id'],columns=['title'],values='rating')  
movieRatings.head(3)
```

Rellenamos los valores no numéricos con ceros.

```python
df = movieRatings.fillna(0)
df.head(3)
```

```python
columna_seleccionada = df['101 Dalmatians (1996)']
columna_seleccionada
```

```python
type(columna_seleccionada)
```

```python
columna_seleccionada_np = columna_seleccionada.to_numpy()
```

```python
type(columna_seleccionada_np)
```

Elevamos cada valor de la columna al número Pi

```python
np.power(columna_seleccionada_np, math.pi)
```
